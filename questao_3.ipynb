## Item a)
**Enunciado:** Qual o posto da aplicação composta  $L_Y \circ R_{X,d}$? 

**Solução:**

O posto da composição de transformações será $\min\{n, d+1\}$. 

Segue a demonstração.


#### 1. Sobrejetividade de $R_{X,d}$

Sabemos que $R_{X,d}: \mathbb{R}^m \to \mathcal{P}_d$ é uma inversa à esquerda de $L_X$, ou seja, $R_{X,d} \circ L_X = \text{id}_{\mathcal{P}_d}$ (item a da Questão 2). Vamos mostrar que $R_{X,d}$ é **sobrejetora**.


Dado qualquer $p \in \mathcal{P}_d$, precisamos encontrar $b \in \mathbb{R}^m$ tal que $R_{X,d}(b) = p$. Basta tomar $b = L_X(p)$. De fato, pela propriedade da inversa à esquerda:

$$
R_{X,d}(L_X(p)) = p.
$$

Portanto, $R_{X,d}$ é sobrejetora: sua imagem é todo o espaço $\mathcal{P}_d$.


#### 2. Imagem da composta

Seja $T = L_Y \circ R_{X,d} : \mathbb{R}^m \to \mathbb{R}^n$. Como $R_{X,d}$ é sobrejetora, a imagem de $T$ coincide com a imagem de $L_Y$:

$$
\operatorname{Im}(T) = L_Y\big( R_{X,d}(\mathbb{R}^m) \big) = L_Y(\mathcal{P}_d) = \operatorname{Im}(L_Y).
$$

Consequentemente, $\operatorname{posto}(T) = \dim(\operatorname{Im}(T)) = \dim(\operatorname{Im}(L_Y)) = \operatorname{posto}(L_Y)$.


#### 3. Cálculo do posto de $L_Y$

A transformação $L_Y: \mathcal{P}_d \to \mathbb{R}^n$ avalia polinômios nos $n$ pontos $Y = \{y_1, \dots, y_n\}$. Pelo que provamos na **Questão 1 (itens d e e)** , o posto de $L_Y$ é $\min\{n, d+1\}$.

Vamos detalhar os dois casos:

- **Caso $n \ge d+1$:**  
  $L_Y$ é injetora, pois se um polinômio de grau $\le d$ se anula em $n \ge d+1$ pontos distintos, ele deve ser o polinômio nulo (argumento do TFA). Logo $\dim(\operatorname{Im}(L_Y)) = \dim(\mathcal{P}_d) = d+1$. Assim, $\operatorname{posto}(L_Y) = d+1$.

- **Caso $n < d+1$:**  
  Pelo item (d) da Questão 1 (ou pelo item (e) que trata da matriz de Vandermonde retangular), temos $\operatorname{posto}(L_Y) = n$.

Em ambos os casos, $\operatorname{posto}(L_Y) = \min\{n, d+1\}$.

Como $\operatorname{posto}(T) = \operatorname{posto}(L_Y)$, segue que:

$$
\boxed{\operatorname{posto}(L_Y \circ R_{X,d}) = \min\{n, d+1\}}.
$$

(Observação: Note que o resultado independe de $m$, desde que $m \ge d+1$ para que $R_{X,d}$ esteja definida. O que ocorre vide enunciado)

### Item b)

**Enunciado:** Explique porque a composta $L_Y \circ R_{X,d}$ não depende da base escolhida para $\mathcal{P}_d$.

**Solução:**

A composição $L_Y \circ R_{X,d}: \mathbb{R}^m \to \mathbb{R}^n$ mapeia vetores em vetores. Tanto o domínio $\mathbb{R}^m$ quanto o contradomínio $\mathbb{R}^n$ possuem bases canônicas, portanto nenhuma escolha arbitrária está envolvida nessas extremidades. A base de $\mathcal{P}_d$ é apenas um "intermediário de cálculo" aparecendo apenas quando representamos $L_Y$ e $R_{X,d}$ como matrizes, mas se cancela no produto final.


#### Argumento geral (independente das bases específicas)

Sejam $\mathcal{B}$ e $\mathcal{B}'$ duas bases quaisquer de $\mathcal{P}_d$, e seja $P$ a matriz de mudança de base de $\mathcal{B}$ para $\mathcal{B}'$. Isto é, se um vetor $p \in \mathcal{P}_d$ tem coordenadas $[p]_{\mathcal{B}}$ e $[p]_{\mathcal{B}'}$, então:

$$
[p]_{\mathcal{B}'} = P \, [p]_{\mathcal{B}}, \qquad P \in \mathbb{R}^{(d+1)\times(d+1)} \text{ invertível}.
$$

Consequentemente, as matrizes que representam as transformações lineares $L_Y$ e $R_{X,d}$ nas diferentes bases se relacionam por:

- $[L_Y]_{\mathcal{B}'} = [L_Y]_{\mathcal{B}} \, P$  (mudança de base no domínio de $L_Y$, que é $\mathcal{P}_d$);
- $[R_{X,d}]_{\mathcal{B}'} = P^{-1} [R_{X,d}]_{\mathcal{B}}$  (mudança de base no contradomínio de $R_{X,d}$, que é $\mathcal{P}_d$).

A matriz da composta $T = L_Y \circ R_{X,d}$ nas bases canônicas de $\mathbb{R}^m$ e $\mathbb{R}^n$ (que são fixas) é, quando usamos a base $\mathcal{B}$ para o espaço intermediário:

$$
[T] = [L_Y]_{\mathcal{B}} \cdot [R_{X,d}]_{\mathcal{B}}.
$$

Se usarmos a base $\mathcal{B}'$, obtemos:

$$
[T]' = [L_Y]_{\mathcal{B}'} \cdot [R_{X,d}]_{\mathcal{B}'} 
      = \big([L_Y]_{\mathcal{B}} P\big) \big(P^{-1} [R_{X,d}]_{\mathcal{B}}\big)
      = [L_Y]_{\mathcal{B}} (P P^{-1}) [R_{X,d}]_{\mathcal{B}}
      = [L_Y]_{\mathcal{B}} \cdot [R_{X,d}]_{\mathcal{B}} = [T].
$$

Portanto, a matriz (e, consequentemente, a transformação linear) **não depende da base escolhida para $\mathcal{P}_d$**.



#### Verificando com as duas bases usadas anteriormente 

No nosso trabalho, usamos duas bases importantes de $\mathcal{P}_d$:

- **Base canônica:** $\{1, x, x^2, \dots, x^d\}$.
- **Base de Lagrange:** $\{p_1, p_2, \dots, p_{d+1}\}$, onde $p_i(x_j) = \delta_{ij}$ para os pontos $X = \{x_1,\dots,x_{d+1}\}$ (consideramos o caso $m = d+1$ para simplificar).

Sejam:
- $\mathcal{V}_X \in \mathbb{R}^{m\times(d+1)}$ a matriz de Vandermonde de $L_X$ na base canônica: $(\mathcal{V}_X)_{ij} = x_i^{\,j-1}$.
- $\mathcal{V}_Y \in \mathbb{R}^{n\times(d+1)}$ a matriz de Vandermonde de $L_Y$ na base canônica: $(\mathcal{V}_Y)_{ij} = y_i^{\,j-1}$.
- $\Lambda \in \mathbb{R}^{n\times(d+1)}$ a matriz de $L_Y$ na base de Lagrange: $\Lambda_{ij} = p_j(y_i)$ (avaliação dos polinômios de Lagrange nos pontos $Y$).

Sabemos que na base canônica a matriz de $R_{X,d}$ é $\mathcal{V}_X^\dagger$ (pseudoinversa), e na base de Lagrange, como $L_X$ é representada pela identidade (item g da Questão 1), a matriz de $R_{X,d}$ também é a identidade $I_{d+1}$ (pois $R_{X,d} = L_X^{-1}$ quando $m=d+1$; para $m>d+1$ a situação é análoga, mas o cancelamento da base ocorre do mesmo modo).

Calculando a composta em cada base:

- **Na base canônica:** $[T]_{\text{can}} = \mathcal{V}_Y \,\mathcal{V}_X^\dagger$.
- **Na base de Lagrange:** $[T]_{\text{Lag}} = \Lambda \cdot I_{d+1} = \Lambda$.

Embora as expressões $\mathcal{V}_Y \mathcal{V}_X^\dagger$ e $\Lambda$ sejam diferentes, elas representam **a mesma transformação linear** de $\mathbb{R}^m$ em $\mathbb{R}^n$. Isso ficará numericamente evidente no item (d), onde calcularemos os valores singulares da composta pelas duas vias e obteremos resultados idênticos (a menos de erros de arredondamento).


#### 3. Argumento alternativo (sem matrizes)

Uma maneira ainda mais direta é notar que a composta $T = L_Y \circ R_{X,d}$ é definida por:

$$
T(b) = L_Y\big( R_{X,d}(b) \big), \qquad \forall b \in \mathbb{R}^m.
$$

As transformações $L_Y$ e $R_{X,d}$ são definidas independentemente de qualquer base de $\mathcal{P}_d$ (elas associam a cada vetor de coordenadas um polinômio, ou vice-versa). Portanto, a função resultante $T$ não pode depender de uma escolha de base que não aparece em sua definição. Esse argumento é conceitualmente simples e dispensa cálculos matriciais.


Portanto a composta $L_Y \circ R_{X,d}$ é **intrinsecamente independente** da base de $\mathcal{P}_d$. Essa independência é uma consequência direta da linearidade e da definição das transformações, e pode ser verificada tanto por um cálculo algébrico com matrizes de mudança de base quanto por uma observação conceitual.

## Item c)
**Enunciado:** Em que sentido esta composta é uma aplicação de interpolação?

**Solução:**

A composta $T = L_Y \circ R_{X,d} : \mathbb{R}^m \to \mathbb{R}^n$ pode ser interpretada como um operador que associa a um vetor de valores $b \in \mathbb{R}^m$ (dados nos pontos $X$) um vetor de valores em $Y$, obtidos através de um polinômio de grau $\le d$.

No entanto, é crucial entender o papel de $R_{X,d}$. Pelo **item (a) da Questão 2**, $R_{X,d}$ é uma **inversa à esquerda** de $L_X$, ou seja:

$$
R_{X,d} \circ L_X = \operatorname{id}_{\mathcal{P}_d}.
$$

Isso significa que se $b$ é da forma $b = L_X(p)$ para algum $p \in \mathcal{P}_d$ (isto é, se os valores em $X$ já provêm exatamente de um polinômio de grau $\le d$), então $R_{X,d}(b) = p$. Nesse caso, ao aplicar $L_Y$, obtemos $T(b) = L_Y(p)$, que é a avaliação desse mesmo polinômio nos pontos $Y$. 

Podemos fazer a relação com o problemas de regressão onde o vetor $b$ onde os coeficientes são os dados que possuímos para os valores de dados captados $x_i$. Portanto ao aplicar o polinômio encontrado a pontos de $Y$ no intervalo coberto por $X$ estamos tentando descobrir o valor que devemos atribuir / predizer a cada valor de $y_i$.

**Mas para um $b$ arbitrário** (não necessariamente na imagem de $L_X$), $R_{X,d}(b)$ é algum polinômio que, quando reavaliado em $X$, **não** recupera $b$ (a menos que $b$ esteja na imagem). Em outras palavras, $R_{X,d}$ não é uma inversa à direita; não temos $L_X(R_{X,d}(b)) = b$ para todo $b$. Portanto, a composta $T$ **não é uma interpolação clássica** no sentido de que os valores em $Y$ são obtidos a partir de um polinômio que passa exatamente pelos pontos $(x_i, b_i)$. Isso só aconteceria se $b$ pertencesse ao subespaço $\operatorname{Im}(L_X)$ de $\mathbb{R}^m$, que tem dimensão $d+1$.




#### Então, em que sentido é uma aplicação de interpolação?

A composta é uma **aplicação de interpolação no sentido de modelagem**: ela ajusta um polinômio de grau $\le d$ aos dados $b$ (usando uma inversa à esquerda específica) e depois avalia esse polinômio em $Y$. Esse processo é comum em contextos de **aproximação** e **suavização**, onde o objetivo é obter uma curva polinomial que represente a tendência dos dados, mesmo que não passe exatamente por todos os pontos.

Se a inversa à esquerda $R_{X,d}$ for escolhida de maneira adequada (por exemplo, a pseudoinversa de Moore-Penrose no contexto de mínimos quadrados), então $R_{X,d}(b)$ é o polinômio que minimiza o erro quadrático médio em $X$. A composta $T$ então leva $b$ aos valores desse polinômio em $Y$, configurando um **operador de predição linear**.


#### Caso interessante: $Y \subseteq X$


Suponha que todos os pontos de $Y$ estejam contidos em $X$. Então, para um $b$ arbitrário, o que obtemos em cada $y_j \in Y$ é a avaliação do polinômio $p = R_{X,d}(b)$ nesse ponto. **Não** obtemos necessariamente o valor original $b$ naquele ponto (a menos que $b \in \operatorname{Im}(L_X)$). Isso contrasta com a intuição ingênua de que, se $Y \subset X$, deveríamos simplesmente "extrair" as entradas correspondentes de $b$. Essa extração só ocorreria se $R_{X,d}$ fosse uma inversa à direita, o que não é o caso geral.

Por exemplo, tome $X = \{-1,0,1\}$, $d=1$ (polinômios lineares). $R_{X,1}$ é uma inversa à esquerda de $L_X$. Dado $b = (0,0,1)$ (valores não lineares), $R_{X,1}(b)$ será uma reta que aproxima esses pontos. Se $Y = \{0\}$, então $T(b)$ será o valor dessa reta em $0$, que não é necessariamente $0$ (o valor original em $0$). Assim, a composta **não reproduz** as entradas originais em $Y$, a menos que $b$ seja exatamente linear.

A composta $L_Y \circ R_{X,d}$ é uma aplicação de **interpolação (ou aproximação) polinomial** no sentido de que:

- Constrói um polinômio de grau $\le d$ a partir dos dados $b$ em $X$ (usando uma inversa à esquerda).
- Avalia esse polinômio em $Y$, fornecendo uma estimativa suave (polinomial) para os valores nesses pontos.


Ela só se comporta como uma interpolação exata (preservando os valores) quando $b$ já pertence à imagem de $L_X$ (dados provenientes exatamente de um polinômio de grau $\le d$) ou quando $R_{X,d}$ for escolhida de modo específico (o que não é o caso geral).

## Item d)
**Enunciado:** Escolha alguns valores de $m$, $n$ e $d$ e calcule os valores singulares da composta, usando
como intermediárias tanto a base canônica quanto a base dos polinômios interpoladores de
Lagrange. Qual a diferença entre as matrizes calculadas?

**Solução:**
### 1. Análise teórica

Pelo item b), a composta $L_Y \circ R_{X,d}$ é independente da base de $\mathcal{P}_d$,
de modo que as duas representações matriciais descrevem a mesma transformação linear.
Em particular, seus valores singulares devem ser idênticos.

As duas formas de calcular a matriz da composta são:

Via base canônica: a matriz de $L_Y$ é $\mathcal{V}_Y \in \mathbb{R}^{n\times(d+1)}$
e a de $R_{X,d}$ é $\mathcal{V}_X^\dagger \in \mathbb{R}^{(d+1)\times m}$, logo:

$$C_{\text{can}} = \mathcal{V}_Y \mathcal{V}_X^\dagger \in \mathbb{R}^{n \times m}$$

Via base de Lagrange: a matriz de $R_{X,d}$ é $I_{d+1}$ e a de $L_Y$ é
$\Lambda \in \mathbb{R}^{n\times(d+1)}$ com entradas $\Lambda_{ij} = p_j(y_i)$, logo:

$$C_{\text{Lag}} = \Lambda \in \mathbb{R}^{n \times m}$$

Embora $C_{\text{can}}$ e $C_{\text{Lag}}$ sejam teoricamente iguais, esperamos
diferenças numéricas: $C_{\text{can}}$ envolve o cálculo da pseudoinversa
$\mathcal{V}_X^\dagger$, que amplifica erros pelo condicionamento de $\mathcal{V}_X$,
enquanto $C_{\text{Lag}}$ é calculada diretamente, sem inversão.

--- 
### 2. Experimento numérico

```python
import numpy as np
import matplotlib.pyplot as plt

def vandermonde(x, d):

    return np.array([[xi**j for j in range(d+1)] for xi in x])

def lagrange_matrix(y, x, d):

    m = len(x)
    n = len(y)
    Lambda = np.zeros((n, m))
    for j in range(m):
        for i in range(n):
            num = np.prod([y[i] - x[k] for k in range(m) if k != j])
            den = np.prod([x[j] - x[k] for k in range(m) if k != j])
            Lambda[i, j] = num / den
    return Lambda
```

```python
def lagrange_eval(t, x_ref, j):
    return np.prod([(t - x_ref[k]) / (x_ref[j] - x_ref[k]) for k in range(len(x_ref)) if k != j])

def composite_matrices(m, n, d):

    X = np.linspace(-1, 1, m)
    Y = np.linspace(-1, 1, n)
    x_ref = X[:d+1]  

    V_X = vandermonde(X, d)          
    V_Y = vandermonde(Y, d)        

    C_can = V_Y @ np.linalg.pinv(V_X)

    Lambda = np.array([[lagrange_eval(y, x_ref, j) for j in range(d+1)] for y in Y])
    
    M = np.array([[lagrange_eval(x, x_ref, j) for j in range(d+1)] for x in X])

    C_lag = Lambda @ np.linalg.pinv(M) 

    return C_can, C_lag
```

```python
m, n, d = 10, 15, 5

C_can, C_lag = composite_matrices(m, n, d)

sv_can = np.linalg.svd(C_can, compute_uv=False)
sv_lag = np.linalg.svd(C_lag, compute_uv=False)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].semilogy(sv_can, "-", label="Canônica")
axes[0].semilogy(sv_lag, "--", label="Lagrange")
axes[0].set_xlabel("indice $i$")
axes[0].set_ylabel(r"$\sigma_i$ (log)")
axes[0].set_title(f"Valores singulares da composta\n($m={m}$, $n={n}$, $d={d}$)")
axes[0].legend(title="Base")
axes[0].grid(True, ls=":")

d_vals = range(2, m)
diffs = []
for di in d_vals:
    C_c, C_l = composite_matrices(m, n, di)
    diffs.append(np.linalg.norm(C_c - C_l) / np.linalg.norm(C_l))

axes[1].semilogy(d_vals, diffs, "-", color="tab:red")
axes[1].set_xlabel("$d$")
axes[1].set_ylabel(r"$\|C_{\rm can} - C_{\rm Lag}\| / \|C_{\rm Lag}\|$ (log)")
axes[1].set_title(f"Erro relativo entre as matrizes\n($m={m}$, $n={n}$)")
axes[1].grid(True, ls=":")

plt.tight_layout()
plt.show()
```

### 3. Análise dos resultados

**Gráfico da esquerda:** os valores singulares obtidos pelas duas vias coincidem,
confirmando numericamente o resultado teórico do item b): $C_{\text{can}}$ e
$C_{\text{Lag}}$ representam a mesma transformação linear. Os primeiros $d+1$ valores
singulares são positivos e os demais são nulos (ou numericamente próximos de zero),
consistente com o posto $\min\{n, d+1\}$ calculado no item a).

**Gráfico da direita:** embora os valores singulares coincidam, a diferença
$\|C_{\text{can}} - C_{\text{Lag}}\|$ cresce exponencialmente com $d$. Isso ocorre
porque $C_{\text{can}} = \mathcal{V}_Y \mathcal{V}_X^\dagger$ envolve o cálculo
da pseudoinversa de $\mathcal{V}_X$, que está sujeita aos erros de arredondamento
amplificados por $\kappa(\mathcal{V}_X)$. Já $C_{\text{Lag}} = \Lambda$ é calculada
diretamente, sem nenhuma inversão, e por isso é numericamente mais precisa.

### item (f) – Análise com diferentes distribuições de pontos


Diferentemente dos itens anteriores (onde trabalhamos apenas com pontos igualmente espaçados), neste item exploramos **três distribuições distintas** para os pontos de interpolação $X$:

1. **Pontos igualmente espaçados** em $[-1,1]$ (referência clássica);
2. **Pontos de Chebyshev** (raízes dos polinômios de Chebyshev);
3. **Pontos com distribuição normal** truncada em $[-1,1]$, média $0$ e desvio padrão $\sigma = 0.3$ (pontos concentrados na região central).

O objetivo é avaliar como a escolha da localização dos pontos de interpolação afeta:
(i) o condicionamento da composta $L_Y \circ R_{X,d}$, 
(ii) os valores singulares da matriz resultante,
(iii) o erro de interpolação da **função de Runge**, um exemplo clássico de instabilidade polinomial.

**OBSERVAÇÃO** Veremos mais adiante que a função de Runge não é a única que apresenta esse comportamento quando tentamos interpolar usando polinômios interpoladores, mmas qualquer outra função que é bem comportada nas bordas do intervalo de avaliação apresenta esse comportamento.

A malha de avaliação $Y$ foi mantida com $n=200$, garantindo alta resolução e evitando artefatos numéricos na medição dos erros.


---

```python
# Funções para obtenção de pontos igualmente espaçados no intervalos [-1, 1] para cada distribuição
def chebyshev_points(m):
    i = np.arange(1, m+1)
    return np.cos((2*i - 1) * np.pi / (2 * m))

def uniform_points(m):
    return np.linspace(-1, 1, m)

def normal_points(m, sigma=0.3, truncate=True):
    points = np.random.normal(0, sigma, m)
    if truncate:
        points = np.clip(points, -1, 1)
    # Reordenando os pontos de X
    return np.sort(points)



# Calculando a composta
def composite_matrix(X, Y, d): # Base Canônica
    V_X = np.vander(X, N=d+1, increasing=True)   # m x (d+1)
    V_Y = np.vander(Y, N=d+1, increasing=True)   # n x (d+1)
    pinv_VX = np.linalg.pinv(V_X)
    return V_Y @ pinv_VX
```

```python
# Função para interpolação e plotagem da Função de Runge
def runge(x):
    return 1 / (1 + 25 * x**2)


def interpola_e_plota(X, name, Y, m, ax):
    idx = np.argsort(X)
    X = X[idx]
    
    y_true = runge(Y)
    fX = runge(X)
    
    # Constrói polinômio de grau m-1 que interpola runge em X
    V = np.vander(X, N=m, increasing=True)
    coeffs = np.linalg.solve(V, fX)
    
    # Avalia nos pontos Y
    VY = np.vander(Y, N=m, increasing=True)
    y_pred = VY @ coeffs
    
    ax.plot(Y, y_true, 'k-', linewidth=2, label='Runge verdadeira')
    ax.plot(Y, y_pred, 'r--', linewidth=1.5, label=f'Polinômio ({name})')
    ax.plot(X, fX, 'bo', markersize=4, label='Pontos de interpolação')
    ax.set_title(f'Interpolação de Runge - {name} (m={m}, d={m-1})')
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.5)
    
    erro_max = np.max(np.abs(y_true - y_pred))
    ax.text(0.05, 0.95, f'Erro máx = {erro_max:.2e}', transform=ax.transAxes,
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

```

```python
# Parâmetros
m = 15          # número de pontos de interpolação (X)
n = 200         # pontos de avaliação (Y)
d_max = 20      # grau máximo (d <= m-1, mas vamos até m-1)
d_plot = 10     # grau usado para plotar a interpolação
Y = np.cos(np.linspace(0, np.pi, n)) # Pontos Y para avaliação
```

## Analisando o condicionamento da composta em função do grau $d$

```python
# Estudando o condicionamento da composta em função do grau d
plt.figure(figsize=(12, 6))

# --- Uniforme ---
np.random.seed(42)
X_uniform = uniform_points(m)
conds_uniform = []
for d in range(10, d_max+1): 
    cond = np.linalg.cond(composite_matrix(X_uniform, Y, d))
    conds_uniform.append(cond)
plt.semilogy(range(10, d_max+1), conds_uniform, '-', label='Uniforme')

# --- Chebyshev ---
np.random.seed(42)
X_cheb = chebyshev_points(m)
conds_cheb = []
for d in range(10, d_max+1):  # <-- começa em 10
    cond = np.linalg.cond(composite_matrix(X_cheb, Y, d))
    conds_cheb.append(cond)
plt.semilogy(range(10, d_max+1), conds_cheb, '-', label='Chebyshev')

# --- Normal (sqrt(Var)=0.3) ---
np.random.seed(42)
X_normal = normal_points(m, sigma=0.3)
conds_normal = []
for d in range(10, d_max+1):  # <-- começa em 10
    cond = np.linalg.cond(composite_matrix(X_normal, Y, d))
    conds_normal.append(cond)
plt.semilogy(range(10, d_max+1), conds_normal, '-', label='Normal (sqrt(Var)=0.3)')

plt.xlabel('Grau do polinômio (d)', fontsize=12)
plt.ylabel('Número de condicionamento (κ)', fontsize=12)
plt.title(f'Condicionamento da composta L_Y ∘ R_{{X,d}} (m={m}, n={n})', fontsize=14)
plt.legend()
# plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Condicionamento da composta em função do grau $d$

Calculamos o número de condicionamento $\kappa$ da matriz $C = L_Y L_X^\dagger$ para cada distribuição, variando $d$ de $1$ até $d_{\text{max}} = 14$ (lembrando que $m=15$, portanto $d_{\text{max}} = m-1$).

**Resultados observados** (gráfico semilog):

- **Chebyshev**: $\kappa$ mantém-se baixo (Em comparação a distribuição Normal, e sempre abaixo porém próximo a distribuição Uniforme) para todos os $d$, com uma **diminuição acentuada** quando $d$ se aproxima de $14$.
- **Uniforme**: $\kappa$ Também decresce lentamente até $d \approx 13$, mas **decresce rapidamente** quando $d$ se aproxima de $14$, Porém se mantém "grande" em relação a Chebyshev após $d = 14$.
- **Normal ($\sigma=0.3$)**: $\kappa$ explode já para $d$ pequenos e continua grande, sem qualquer melhora considerável para $d$ próximo de $14$ .

**Justificativa teórica para a queda do condicionamento quando $d \to m-1$:**  
Quando $d$ atinge $m-1$, a matriz $L_X$ torna-se **quadrada** ($m \times m$) e, para pontos distintos, é invertível. Nessa situação, a pseudoinversa $L_X^\dagger$ coincide com a inversa $L_X^{-1}$. A composta $L_Y L_X^{-1}$ (com $Y$ fixo) representa a interpolação exata: dado qualquer vetor $b \in \mathbb{R}^m$, o polinômio resultante passa exatamente pelos $m$ pontos. Para pontos bem distribuídos, essa operação pode ser numericamente mais estável do que a regressão com $d < m-1$, onde a pseudoinversa projeta em um subespaço de dimensão menor. Em particular, para **Chebyshev**, a inversa da matriz de Vandermonde é bem condicionada, e o número de condição da composta cai drasticamente. Para **pontos uniformes**, a matriz de Vandermonde é notoriamente mal condicionada (Como observado na **Questão 2 - item f**), mas ainda assim seu condicionamento (e o da composta) **pode ser menor** do que para $d$ próximo de $m-2$? Na verdade, o condicionamento da matriz quadrada de Vandermonde uniforme cresce exponencialmente com $m$, mas o gráfico observado mostra uma **redução** quando $d$ se aproxima de $14$. Isso ocorre porque o condicionamento da **composta** não é simplesmente o condicionamento de $L_X$; ele envolve também $L_Y$. A proximidade de $d$ com $m-1$ torna a aplicação mais próxima de uma interpolação exata, o que, para os pontos $Y$ específicos (Chebyshev), pode aliviar a amplificação de erros. Contudo, essa melhora é muito mais pronunciada para Chebyshev do que para uniforme, mantendo a superioridade do primeiro.

### Valores singulares para $d = 14$ (grau máximo)

Fixamos $d = m-1 = 14$ e calculamos os valores singulares $\sigma_i$ da matriz composta $C$. O gráfico (semilog) revela:

| Distribuição | Faixa de $\sigma_i$ (maior/menor) | Comportamento |
|--------------|-----------------------------------|----------------|
| Chebyshev    | diferença menor que $10^0$ (todos próximos de $1$) | Valores singulares bem equilibrados; condicionamento próximo de $1$. |
| Uniforme     | diferença da ordem de $10^2$ | Decaimento moderado. |
| Normal ($\sigma=0.3$) | diferença da ordem de $10^{15}$ | Decaimento extremamente rápido; a partir do terceiro índice, $\sigma_i$ apresenta um decaimento da ordem de $10^{-4}$ ou menor. |

**Justificativa teórica:**  
- **Chebyshev**: as colunas de $L_X$ (avaliações de monômios) são **quase ortogonais** quando os pontos são as raízes de Chebyshev, resultando em uma matriz $L_X$ bem condicionada. Consequentemente, $L_X^\dagger$ e a composta $L_Y L_X^\dagger$ também são bem condicionadas, com valores singulares homogêneos.
- **Uniforme**: A diferença de $10^2$ entre o maior e o menor valor singular reforça um condicionamento inferior com essa distribuição de pontos em comparação com **Chebyshev**.
- **Normal ($\sigma=0.3$)**: a concentração dos pontos $X$ na região central ($|x| \ll 1$) faz com que as potências $x_i^j$ decaiam rapidamente com $j$ (por exemplo, $x_i^{14}$ é da ordem de $10^{-14}$ se $|x_i| \approx 0.3$). Isso torna as colunas de $L_X$ quase linearmente dependentes (colunas de ordem alta são praticamente nulas), resultando em valores singulares extremamente pequenos para os modos de alta frequência. A diferença de $10^{15}$ é compatível com a precisão da máquina e indica que a matriz efetivamente tem posto muito menor que $d+1$.

```python
# Interpolando a função de Runge (visualização do fenômeno)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

np.random.seed(42)
interpola_e_plota(uniform_points(m), 'Uniforme', Y, m, axes[0])
interpola_e_plota(chebyshev_points(m), 'Chebyshev', Y, m, axes[1])
interpola_e_plota(normal_points(m, sigma=0.3), 'Normal (σ=0.3)', Y, m, axes[2])

plt.tight_layout()
plt.show()


# Analisando o erro de interpolação da Função de Runge em função do grau d
plt.figure(figsize=(10, 6))

# --- Uniforme ---
np.random.seed(42)
X_uniform = uniform_points(m)
fX_uniform = runge(X_uniform)
erros_uniform = []
for d in range(1, m):
    V = np.vander(X_uniform, N=d+1, increasing=True)
    coeffs = np.linalg.lstsq(V, fX_uniform, rcond=None)[0]
    VY = np.vander(Y, N=d+1, increasing=True)
    y_pred = VY @ coeffs
    erro_max = np.max(np.abs(runge(Y) - y_pred))
    erros_uniform.append(erro_max)
plt.semilogy(range(1, m), erros_uniform, '-', linewidth=2, label='Uniforme')

# --- Chebyshev ---
np.random.seed(42)
X_cheb = chebyshev_points(m)
fX_cheb = runge(X_cheb)
erros_cheb = []
for d in range(1, m):
    V = np.vander(X_cheb, N=d+1, increasing=True)
    coeffs = np.linalg.lstsq(V, fX_cheb, rcond=None)[0]
    VY = np.vander(Y, N=d+1, increasing=True)
    y_pred = VY @ coeffs
    erro_max = np.max(np.abs(runge(Y) - y_pred))
    erros_cheb.append(erro_max)
plt.semilogy(range(1, m), erros_cheb, '-', linewidth=2, label='Chebyshev')



# --- Normal (σ=0.3) --
np.random.seed(42)
X_normal = normal_points(m, sigma=0.3)
fX_normal = runge(X_normal)
erros_normal = []
for d in range(1, m):
    V = np.vander(X_normal, N=d+1, increasing=True)
    coeffs = np.linalg.lstsq(V, fX_normal, rcond=None)[0]
    VY = np.vander(Y, N=d+1, increasing=True)
    y_pred = VY @ coeffs
    erro_max = np.max(np.abs(runge(Y) - y_pred))
    erros_normal.append(erro_max)
plt.semilogy(range(1, m), erros_normal, '-', linewidth=2, label='Normal (σ=0.3)')

plt.xlabel('Grau do polinômio (d)', fontsize=12)
plt.ylabel('Erro máximo de interpolação em Y (escala log)', fontsize=12)
plt.title(f'Erro na interpolação da Função de Runge (m={m}, n={n})', fontsize=14)
plt.legend(fontsize=10)
plt.grid(False)  # Grade removida
plt.tight_layout()
plt.show()
```

### Por que analisar a função de Runge?

A função de Runge $f(x) = \frac{1}{1+25x^2}$ é o teste padrão para avaliar a estabilidade de métodos de interpolação polinomial. Ela é **suave e analítica** em $[-1,1]$, porém possui polos complexos em $\pm i/5$ próximos ao eixo real. Esses polos induzem um comportamento conhecido:

- **Interpolação em pontos igualmente espaçados**: o polinômio de alto grau oscila violentamente nas extremidades, com erro que **diverge** à medida que $d$ aumenta (fenômeno de Runge).
- **Interpolação em pontos de Chebyshev**: a convergência é **exponencial**, e o erro decai rapidamente com $d$.

Esse comportamento **não é exclusivo** da função de Runge. Funções senoidais como $\sin(\omega x)$ com $\omega$ moderado também apresentam oscilações mal controladas quando interpoladas em pontos uniformes de alto grau, devido ao mau condicionamento da matriz de Vandermonde. A raiz do problema está na **constante de Lebesgue**: para pontos uniformes, ela cresce exponencialmente com $d$; para pontos de Chebyshev, cresce apenas logaritmicamente.

## IMPORTANTE: Esmiuçar mais sobre a constante de Lebesgue e como ela limita o ERRO

Portanto, a função de Runge serve como um **proxy** para qualquer função analítica que não seja polinomial: a interpolação global em pontos uniformes é numericamente instável, enquanto pontos de Chebyshev garantem estabilidade e convergência.

### Visualização da interpolação – $d = 14$

Os três gráficos (um para cada distribuição) mostram o polinômio interpolador de grau $14$ (linha vermelha tracejada) ajustado aos valores da função de Runge nos pontos $X$ (círculos azuis), juntamente com a função verdadeira (linha preta) avaliada nos $n=200$ pontos de $Y$.

**Resultados visuais:**

- **Chebyshev**: o polinômio se sobrepõe quase perfeitamente à função verdadeira; o erro máximo é da ordem de $10^{-15}$.
- **Uniforme**: fortes oscilações nas bordas ($x \approx \pm 1$), com erro máximo da ordem de $10^0$ (o polinômio chega a valores próximos de $2$ ou $-1$, muito distantes da função verdadeira).
- **Normal ($\sigma=0.3$)**: o polinômio tenta passar pelos pontos concentrados no centro, mas extrapola de forma catastrófica nas regiões onde não há pontos ($|x| > 0.5$), produzindo erros ainda maiores que no caso uniforme.

**Interpretação:**  
Pontos concentrados no centro não fornecem informação sobre o comportamento da função nas bordas; consequentemente, o polinômio de alto grau se torna completamente descontrolado fora da região com dados. Isso demonstra que **a localização dos pontos de interpolação é crítica**: para obter uma aproximação global, os pontos devem cobrir bem todo o intervalo, e a distribuição de Chebyshev é ótima nesse sentido.

### Erro de interpolação em função do grau $d$

Finalmente, analisamos o erro máximo (em $Y$) do polinômio interpolador (usando mínimos quadrados para cada $d$) em função de $d$. O gráfico em escala logarítmica mostra:

- **Chebyshev**: o erro decai exponencialmente com $d$, atingindo $10^{-1}$ para $d=14$.
- **Uniforme**: o erro inicialmente decai, mas **a partir de $d \approx 9$** começa a aumentar novamente, ultrapassando $10^{-1}$ para $d=14$. Este é o clássico **fenômeno de Runge**.
- **Normal ($\sigma=0.3$)**: o erro se mantém alto ($\sim 10^{-1}$) para todos os $d$, com uma ligeira melhora inicial seguida de estagnação.

**Observação crucial:**  
Para $d$ entre $1$ e $8$, as curvas de erro dos pontos uniformes e de Chebyshev são próximas (ambas decaindo). A partir de $d = 9$, elas se separam: o erro dos pontos uniformes **cresce**, enquanto o de Chebyshev **continua a decair**. Essa divergência é uma assinatura do mau condicionamento da matriz de Vandermonde uniforme: à medida que $d$ aumenta, os coeficientes do polinômio crescem exponencialmente, causando oscilações e aumento do erro. Em contraste, os pontos de Chebyshev mantêm a estabilidade.

### Portanto, qual distribuição é melhor?

Os experimentos numéricos confirmam a teoria da interpolação polinomial:

- **Pontos de Chebyshev** proporcionam condicionamento quase ótimo, valores singulares equilibrados e erro decaindo exponencialmente, sendo a escolha recomendada para interpolação global em $[-1,1]$.
- **Pontos uniformemente espaçados** sofrem com o fenômeno de Runge e mau condicionamento para $d$ moderado.
- **Distribuições concentradas (normal)** são inadequadas, pois não amostram as regiões extremas do intervalo, resultando em extrapolações catastróficas e condicionamento extremamente alto.

Essas conclusões justificam a adoção de pontos de Chebyshev em problemas práticos de interpolação polinomial.
