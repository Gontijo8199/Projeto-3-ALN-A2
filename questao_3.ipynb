## Item a)
**Enunciado:** Qual o posto da aplicação composta  $L_Y \circ R_{X,d}$? 

**Solução:**

O posto da composição de transformações será $\min\{n, d+1\}$. 

Segue a demonstração.


#### 1. Sobrejetividade de $R_{X,d}$

Sabemos que $R_{X,d}: \mathbb{R}^m \to \mathcal{P}_d$ é uma inversa à esquerda de $L_X$, ou seja, $R_{X,d} \circ L_X = \text{id}_{\mathcal{P}_d}$ (item a da Questão 2). Vamos mostrar que $R_{X,d}$ é **sobrejetora**.


Dado qualquer $p \in \mathcal{P}_d$, precisamos encontrar $b \in \mathbb{R}^m$ tal que $R_{X,d}(b) = p$. Basta tomar $b = L_X(p)$. De fato, pela propriedade da inversa à esquerda:

$$
R_{X,d}(L_X(p)) = p.
$$

Portanto, $R_{X,d}$ é sobrejetora: sua imagem é todo o espaço $\mathcal{P}_d$.


#### 2. Imagem da composta

Seja $T = L_Y \circ R_{X,d} : \mathbb{R}^m \to \mathbb{R}^n$. Como $R_{X,d}$ é sobrejetora, a imagem de $T$ coincide com a imagem de $L_Y$:

$$
\operatorname{Im}(T) = L_Y\big( R_{X,d}(\mathbb{R}^m) \big) = L_Y(\mathcal{P}_d) = \operatorname{Im}(L_Y).
$$

Consequentemente, $\operatorname{posto}(T) = \dim(\operatorname{Im}(T)) = \dim(\operatorname{Im}(L_Y)) = \operatorname{posto}(L_Y)$.


#### 3. Cálculo do posto de $L_Y$

A transformação $L_Y: \mathcal{P}_d \to \mathbb{R}^n$ avalia polinômios nos $n$ pontos $Y = \{y_1, \dots, y_n\}$. Pelo que provamos na **Questão 1 (itens d e e)** , o posto de $L_Y$ é $\min\{n, d+1\}$.

Vamos detalhar os dois casos:

- **Caso $n \ge d+1$:**  
  $L_Y$ é injetora, pois se um polinômio de grau $\le d$ se anula em $n \ge d+1$ pontos distintos, ele deve ser o polinômio nulo (argumento do TFA). Logo $\dim(\operatorname{Im}(L_Y)) = \dim(\mathcal{P}_d) = d+1$. Assim, $\operatorname{posto}(L_Y) = d+1$.

- **Caso $n < d+1$:**  
  Pelo item (d) da Questão 1 (ou pelo item (e) que trata da matriz de Vandermonde retangular), temos $\operatorname{posto}(L_Y) = n$.

Em ambos os casos, $\operatorname{posto}(L_Y) = \min\{n, d+1\}$.

Como $\operatorname{posto}(T) = \operatorname{posto}(L_Y)$, segue que:

$$
\boxed{\operatorname{posto}(L_Y \circ R_{X,d}) = \min\{n, d+1\}}.
$$

(Observação: Note que o resultado independe de $m$, desde que $m \ge d+1$ para que $R_{X,d}$ esteja definida. O que ocorre vide enunciado)

### Item b)

**Enunciado:** Explique porque a composta $L_Y \circ R_{X,d}$ não depende da base escolhida para $\mathcal{P}_d$.

**Solução:**

A composição $L_Y \circ R_{X,d}: \mathbb{R}^m \to \mathbb{R}^n$ mapeia vetores em vetores. Tanto o domínio $\mathbb{R}^m$ quanto o contradomínio $\mathbb{R}^n$ possuem bases canônicas, portanto nenhuma escolha arbitrária está envolvida nessas extremidades. A base de $\mathcal{P}_d$ é apenas um "intermediário de cálculo" aparecendo apenas quando representamos $L_Y$ e $R_{X,d}$ como matrizes, mas se cancela no produto final.


#### Argumento geral (independente das bases específicas)

Sejam $\mathcal{B}$ e $\mathcal{B}'$ duas bases quaisquer de $\mathcal{P}_d$, e seja $P$ a matriz de mudança de base de $\mathcal{B}$ para $\mathcal{B}'$. Isto é, se um vetor $p \in \mathcal{P}_d$ tem coordenadas $[p]_{\mathcal{B}}$ e $[p]_{\mathcal{B}'}$, então:

$$
[p]_{\mathcal{B}'} = P \, [p]_{\mathcal{B}}, \qquad P \in \mathbb{R}^{(d+1)\times(d+1)} \text{ invertível}.
$$

Consequentemente, as matrizes que representam as transformações lineares $L_Y$ e $R_{X,d}$ nas diferentes bases se relacionam por:

- $[L_Y]_{\mathcal{B}'} = [L_Y]_{\mathcal{B}} \, P$  (mudança de base no domínio de $L_Y$, que é $\mathcal{P}_d$);
- $[R_{X,d}]_{\mathcal{B}'} = P^{-1} [R_{X,d}]_{\mathcal{B}}$  (mudança de base no contradomínio de $R_{X,d}$, que é $\mathcal{P}_d$).

A matriz da composta $T = L_Y \circ R_{X,d}$ nas bases canônicas de $\mathbb{R}^m$ e $\mathbb{R}^n$ (que são fixas) é, quando usamos a base $\mathcal{B}$ para o espaço intermediário:

$$
[T] = [L_Y]_{\mathcal{B}} \cdot [R_{X,d}]_{\mathcal{B}}.
$$

Se usarmos a base $\mathcal{B}'$, obtemos:

$$
[T]' = [L_Y]_{\mathcal{B}'} \cdot [R_{X,d}]_{\mathcal{B}'} 
      = \big([L_Y]_{\mathcal{B}} P\big) \big(P^{-1} [R_{X,d}]_{\mathcal{B}}\big)
      = [L_Y]_{\mathcal{B}} (P P^{-1}) [R_{X,d}]_{\mathcal{B}}
      = [L_Y]_{\mathcal{B}} \cdot [R_{X,d}]_{\mathcal{B}} = [T].
$$

Portanto, a matriz (e, consequentemente, a transformação linear) **não depende da base escolhida para $\mathcal{P}_d$**.



#### Verificando com as duas bases usadas anteriormente 

No nosso trabalho, usamos duas bases importantes de $\mathcal{P}_d$:

- **Base canônica:** $\{1, x, x^2, \dots, x^d\}$.
- **Base de Lagrange:** $\{p_1, p_2, \dots, p_{d+1}\}$, onde $p_i(x_j) = \delta_{ij}$ para os pontos $X = \{x_1,\dots,x_{d+1}\}$ (consideramos o caso $m = d+1$ para simplificar).

Sejam:
- $\mathcal{V}_X \in \mathbb{R}^{m\times(d+1)}$ a matriz de Vandermonde de $L_X$ na base canônica: $(\mathcal{V}_X)_{ij} = x_i^{\,j-1}$.
- $\mathcal{V}_Y \in \mathbb{R}^{n\times(d+1)}$ a matriz de Vandermonde de $L_Y$ na base canônica: $(\mathcal{V}_Y)_{ij} = y_i^{\,j-1}$.
- $\Lambda \in \mathbb{R}^{n\times(d+1)}$ a matriz de $L_Y$ na base de Lagrange: $\Lambda_{ij} = p_j(y_i)$ (avaliação dos polinômios de Lagrange nos pontos $Y$).

Sabemos que na base canônica a matriz de $R_{X,d}$ é $\mathcal{V}_X^\dagger$ (pseudoinversa), e na base de Lagrange, como $L_X$ é representada pela identidade (item g da Questão 1), a matriz de $R_{X,d}$ também é a identidade $I_{d+1}$ (pois $R_{X,d} = L_X^{-1}$ quando $m=d+1$; para $m>d+1$ a situação é análoga, mas o cancelamento da base ocorre do mesmo modo).

Calculando a composta em cada base:

- **Na base canônica:** $[T]_{\text{can}} = \mathcal{V}_Y \,\mathcal{V}_X^\dagger$.
- **Na base de Lagrange:** $[T]_{\text{Lag}} = \Lambda \cdot I_{d+1} = \Lambda$.

Embora as expressões $\mathcal{V}_Y \mathcal{V}_X^\dagger$ e $\Lambda$ sejam diferentes, elas representam **a mesma transformação linear** de $\mathbb{R}^m$ em $\mathbb{R}^n$. Isso ficará numericamente evidente no item (d), onde calcularemos os valores singulares da composta pelas duas vias e obteremos resultados idênticos (a menos de erros de arredondamento).


#### 3. Argumento alternativo (sem matrizes)

Uma maneira ainda mais direta é notar que a composta $T = L_Y \circ R_{X,d}$ é definida por:

$$
T(b) = L_Y\big( R_{X,d}(b) \big), \qquad \forall b \in \mathbb{R}^m.
$$

As transformações $L_Y$ e $R_{X,d}$ são definidas independentemente de qualquer base de $\mathcal{P}_d$ (elas associam a cada vetor de coordenadas um polinômio, ou vice-versa). Portanto, a função resultante $T$ não pode depender de uma escolha de base que não aparece em sua definição. Esse argumento é conceitualmente simples e dispensa cálculos matriciais.


Portanto a composta $L_Y \circ R_{X,d}$ é **intrinsecamente independente** da base de $\mathcal{P}_d$. Essa independência é uma consequência direta da linearidade e da definição das transformações, e pode ser verificada tanto por um cálculo algébrico com matrizes de mudança de base quanto por uma observação conceitual.

## Item c)
**Enunciado:** Em que sentido esta composta é uma aplicação de interpolação?

**Solução:**

A composta $T = L_Y \circ R_{X,d} : \mathbb{R}^m \to \mathbb{R}^n$ pode ser interpretada como um operador que associa a um vetor de valores $b \in \mathbb{R}^m$ (dados nos pontos $X$) um vetor de valores em $Y$, obtidos através de um polinômio de grau $\le d$.

No entanto, é crucial entender o papel de $R_{X,d}$. Pelo **item (a) da Questão 2**, $R_{X,d}$ é uma **inversa à esquerda** de $L_X$, ou seja:

$$
R_{X,d} \circ L_X = \operatorname{id}_{\mathcal{P}_d}.
$$

Isso significa que se $b$ é da forma $b = L_X(p)$ para algum $p \in \mathcal{P}_d$ (isto é, se os valores em $X$ já provêm exatamente de um polinômio de grau $\le d$), então $R_{X,d}(b) = p$. Nesse caso, ao aplicar $L_Y$, obtemos $T(b) = L_Y(p)$, que é a avaliação desse mesmo polinômio nos pontos $Y$.

**Mas para um $b$ arbitrário** (não necessariamente na imagem de $L_X$), $R_{X,d}(b)$ é algum polinômio que, quando reavaliado em $X$, **não** recupera $b$ (a menos que $b$ esteja na imagem). Em outras palavras, $R_{X,d}$ não é uma inversa à direita; não temos $L_X(R_{X,d}(b)) = b$ para todo $b$. Portanto, a composta $T$ **não é uma interpolação clássica** no sentido de que os valores em $Y$ são obtidos a partir de um polinômio que passa exatamente pelos pontos $(x_i, b_i)$. Isso só aconteceria se $b$ pertencesse ao subespaço $\operatorname{Im}(L_X)$ de $\mathbb{R}^m$, que tem dimensão $d+1$.




#### Então, em que sentido é uma aplicação de interpolação?

A composta é uma **aplicação de interpolação no sentido de modelagem**: ela ajusta um polinômio de grau $\le d$ aos dados $b$ (usando uma inversa à esquerda específica) e depois avalia esse polinômio em $Y$. Esse processo é comum em contextos de **aproximação** e **suavização**, onde o objetivo é obter uma curva polinomial que represente a tendência dos dados, mesmo que não passe exatamente por todos os pontos.

Se a inversa à esquerda $R_{X,d}$ for escolhida de maneira adequada (por exemplo, a pseudoinversa de Moore-Penrose no contexto de mínimos quadrados), então $R_{X,d}(b)$ é o polinômio que minimiza o erro quadrático médio em $X$. A composta $T$ então leva $b$ aos valores desse polinômio em $Y$, configurando um **operador de predição linear**.


#### Caso interessante: $Y \subseteq X$


Suponha que todos os pontos de $Y$ estejam contidos em $X$. Então, para um $b$ arbitrário, o que obtemos em cada $y_j \in Y$ é a avaliação do polinômio $p = R_{X,d}(b)$ nesse ponto. **Não** obtemos necessariamente o valor original $b$ naquele ponto (a menos que $b \in \operatorname{Im}(L_X)$). Isso contrasta com a intuição ingênua de que, se $Y \subset X$, deveríamos simplesmente "extrair" as entradas correspondentes de $b$. Essa extração só ocorreria se $R_{X,d}$ fosse uma inversa à direita, o que não é o caso geral.

Por exemplo, tome $X = \{-1,0,1\}$, $d=1$ (polinômios lineares). $R_{X,1}$ é uma inversa à esquerda de $L_X$. Dado $b = (0,0,1)$ (valores não lineares), $R_{X,1}(b)$ será uma reta que aproxima esses pontos. Se $Y = \{0\}$, então $T(b)$ será o valor dessa reta em $0$, que não é necessariamente $0$ (o valor original em $0$). Assim, a composta **não reproduz** as entradas originais em $Y$, a menos que $b$ seja exatamente linear.

A composta $L_Y \circ R_{X,d}$ é uma aplicação de **interpolação (ou aproximação) polinomial** no sentido de que:

- Constrói um polinômio de grau $\le d$ a partir dos dados $b$ em $X$ (usando uma inversa à esquerda).
- Avalia esse polinômio em $Y$, fornecendo uma estimativa suave (polinomial) para os valores nesses pontos.


Ela só se comporta como uma interpolação exata (preservando os valores) quando $b$ já pertence à imagem de $L_X$ (dados provenientes exatamente de um polinômio de grau $\le d$) ou quando $R_{X,d}$ for escolhida de modo específico (o que não é o caso geral).

## Item d)
**Enunciado:** Escolha alguns valores de $m$, $n$ e $d$ e calcule os valores singulares da composta, usando
como intermediárias tanto a base canônica quanto a base dos polinômios interpoladores de
Lagrange. Qual a diferença entre as matrizes calculadas?

**Solução:**
### 1. Análise teórica

Pelo item b), a composta $L_Y \circ R_{X,d}$ é independente da base de $\mathcal{P}_d$,
de modo que as duas representações matriciais descrevem a mesma transformação linear.
Em particular, seus valores singulares devem ser idênticos.

As duas formas de calcular a matriz da composta são:

Via base canônica: a matriz de $L_Y$ é $\mathcal{V}_Y \in \mathbb{R}^{n\times(d+1)}$
e a de $R_{X,d}$ é $\mathcal{V}_X^\dagger \in \mathbb{R}^{(d+1)\times m}$, logo:

$$C_{\text{can}} = \mathcal{V}_Y \mathcal{V}_X^\dagger \in \mathbb{R}^{n \times m}$$

Via base de Lagrange: a matriz de $R_{X,d}$ é $I_{d+1}$ e a de $L_Y$ é
$\Lambda \in \mathbb{R}^{n\times(d+1)}$ com entradas $\Lambda_{ij} = p_j(y_i)$, logo:

$$C_{\text{Lag}} = \Lambda \in \mathbb{R}^{n \times m}$$

Embora $C_{\text{can}}$ e $C_{\text{Lag}}$ sejam teoricamente iguais, esperamos
diferenças numéricas: $C_{\text{can}}$ envolve o cálculo da pseudoinversa
$\mathcal{V}_X^\dagger$, que amplifica erros pelo condicionamento de $\mathcal{V}_X$,
enquanto $C_{\text{Lag}}$ é calculada diretamente, sem inversão.

--- 
### 2. Experimento numérico

```python
import numpy as np
import matplotlib.pyplot as plt

def vandermonde(x, d):

    return np.array([[xi**j for j in range(d+1)] for xi in x])

def lagrange_matrix(y, x, d):

    m = len(x)
    n = len(y)
    Lambda = np.zeros((n, m))
    for j in range(m):
        for i in range(n):
            num = np.prod([y[i] - x[k] for k in range(m) if k != j])
            den = np.prod([x[j] - x[k] for k in range(m) if k != j])
            Lambda[i, j] = num / den
    return Lambda
```

```python
def lagrange_eval(t, x_ref, j):
    return np.prod([(t - x_ref[k]) / (x_ref[j] - x_ref[k]) for k in range(len(x_ref)) if k != j])

def composite_matrices(m, n, d):

    X = np.linspace(-1, 1, m)
    Y = np.linspace(-1, 1, n)
    x_ref = X[:d+1]  

    V_X = vandermonde(X, d)          
    V_Y = vandermonde(Y, d)        

    C_can = V_Y @ np.linalg.pinv(V_X)

    Lambda = np.array([[lagrange_eval(y, x_ref, j) for j in range(d+1)] for y in Y])
    
    M = np.array([[lagrange_eval(x, x_ref, j) for j in range(d+1)] for x in X])

    C_lag = Lambda @ np.linalg.pinv(M) 

    return C_can, C_lag
```

```python
m, n, d = 10, 15, 5

C_can, C_lag = composite_matrices(m, n, d)

sv_can = np.linalg.svd(C_can, compute_uv=False)
sv_lag = np.linalg.svd(C_lag, compute_uv=False)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].semilogy(sv_can, "-", label="Canônica")
axes[0].semilogy(sv_lag, "--", label="Lagrange")
axes[0].set_xlabel("indice $i$")
axes[0].set_ylabel(r"$\sigma_i$ (log)")
axes[0].set_title(f"Valores singulares da composta\n($m={m}$, $n={n}$, $d={d}$)")
axes[0].legend(title="Base")
axes[0].grid(True, ls=":")

d_vals = range(2, m)
diffs = []
for di in d_vals:
    C_c, C_l = composite_matrices(m, n, di)
    diffs.append(np.linalg.norm(C_c - C_l) / np.linalg.norm(C_l))

axes[1].semilogy(d_vals, diffs, "-", color="tab:red")
axes[1].set_xlabel("$d$")
axes[1].set_ylabel(r"$\|C_{\rm can} - C_{\rm Lag}\| / \|C_{\rm Lag}\|$ (log)")
axes[1].set_title(f"Erro relativo entre as matrizes\n($m={m}$, $n={n}$)")
axes[1].grid(True, ls=":")

plt.tight_layout()
plt.show()
```

### 3. Análise dos resultados

**Gráfico da esquerda:** os valores singulares obtidos pelas duas vias coincidem,
confirmando numericamente o resultado teórico do item b): $C_{\text{can}}$ e
$C_{\text{Lag}}$ representam a mesma transformação linear. Os primeiros $d+1$ valores
singulares são positivos e os demais são nulos (ou numericamente próximos de zero),
consistente com o posto $\min\{n, d+1\}$ calculado no item a).

**Gráfico da direita:** embora os valores singulares coincidam, a diferença
$\|C_{\text{can}} - C_{\text{Lag}}\|$ cresce exponencialmente com $d$. Isso ocorre
porque $C_{\text{can}} = \mathcal{V}_Y \mathcal{V}_X^\dagger$ envolve o cálculo
da pseudoinversa de $\mathcal{V}_X$, que está sujeita aos erros de arredondamento
amplificados por $\kappa(\mathcal{V}_X)$. Já $C_{\text{Lag}} = \Lambda$ é calculada
diretamente, sem nenhuma inversão, e por isso é numericamente mais precisa.
