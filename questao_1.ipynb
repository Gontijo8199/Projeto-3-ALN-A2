## Item a)
**Enunciado:** Seja $X = \{x_1, \ldots, x_m\}$ um conjunto de pontos distintos em $\mathbb{R}$. Defina o funcional de avaliação $L_X: \mathcal{F} \to \mathbb{R}^m$ por $L_X(f) = (f(x_1), \ldots, f(x_m))$. Mostre que $L_X$ é uma transformação linear.

**Solução:**

Para mostrar que $L_X$ é uma transformação linear, devemos verificar duas propriedades:


### 1. Preservação da soma

Sejam $f_1, f_2 \in \mathcal{F}$ (funções reais). Então:

$$
\begin{aligned}
L_X(f_1 + f_2) &= \big((f_1 + f_2)(x_1), \ldots, (f_1 + f_2)(x_m)\big) \quad \text{(definição de } L_X\text{)} \\[4pt]
&= \big(f_1(x_1) + f_2(x_1), \ldots, f_1(x_m) + f_2(x_m)\big) \quad \text{(soma de funções reais)} \\[4pt]
&= \big(f_1(x_1), \ldots, f_1(x_m)\big) + \big(f_2(x_1), \ldots, f_2(x_m)\big) \quad \text{(soma em } \mathbb{R}^m\text{)} \\[4pt]
&= L_X(f_1) + L_X(f_2) \quad \text{(definição de } L_X\text{)} 
\end{aligned}
$$

**Primeira propriedade verificada.**


### 2. Preservação da multiplicação por escalar

Sejam $\alpha \in \mathbb{R}$ e $f \in \mathcal{F}$. Então:

$$
\begin{aligned}
L_X(\alpha \cdot f) &= \big((\alpha f)(x_1), \ldots, (\alpha f)(x_m)\big) \quad \text{(definição de } L_X\text{)} \\[4pt]
&= \big(\alpha \cdot f(x_1), \ldots, \alpha \cdot f(x_m)\big) \quad \text{(multiplicação de função por escalar)} \\[4pt]
&= \alpha \cdot \big(f(x_1), \ldots, f(x_m)\big) \quad \text{(escalar em } \mathbb{R}^m\text{)} \\[4pt]
&= \alpha \cdot L_X(f) \quad \text{(definição de } L_X\text{)}
\end{aligned}
$$

**Segunda propriedade verificada.**

Como $L_X$ satisfaz ambas as propriedades, **$L_X$ é uma transformação linear** de $\mathcal{F}$ em $\mathbb{R}^m$.

## Item b)
**Enunciado:** Mostre que $\mathcal{P}_d$ é um subespaço linear de $\mathcal{F}$ e conclua que $L_X$ restrita a $\mathcal{P}_d$ é linear.

**Solução:**

Como $\mathcal{F}$ é um espaço vetorial e $\mathcal{P}_d \subset \mathcal{F}$, basta mostrar que $\mathcal{P}_d$ satisfaz as duas propriedades de subespaço:


### 1. Fechamento sob a soma

Sejam $p_1, p_2 \in \mathcal{P}_d$. Então $\deg(p_1) \leq d$ e $\deg(p_2) \leq d$.

Escrevendo os polinômios na forma explícita:

$$
\begin{aligned}
p_1(x) &= a_{10} + a_{11}x + a_{12}x^2 + \cdots + a_{1d}x^d \\[4pt]
p_2(x) &= a_{20} + a_{21}x + a_{22}x^2 + \cdots + a_{2d}x^d
\end{aligned}
$$

onde $a_{1i}, a_{2i} \in \mathbb{R}$ para todo $i = 0, 1, \ldots, d$.

Somando os polinômios:

$$
(p_1 + p_2)(x) = (a_{10} + a_{20}) + (a_{11} + a_{21})x + (a_{12} + a_{22})x^2 + \cdots + (a_{1d} + a_{2d})x^d
$$

Como $a_{1i}, a_{2i} \in \mathbb{R}$ para cada $i$, segue que $(a_{1i} + a_{2i}) \in \mathbb{R}$.

Além disso:

$$
\deg(p_1 + p_2) \leq \max\{\deg(p_1), \deg(p_2)\} \leq d
$$

Portanto, $(p_1 + p_2) \in \mathcal{P}_d$.

**Primeira propriedade verificada.**


### 2. Fechamento sob multiplicação por escalar

Sejam $\alpha \in \mathbb{R}$ e $p \in \mathcal{P}_d$. Como $p \in \mathcal{P}_d$, temos $\deg(p) \leq d$.

Para o polinômio $\alpha \cdot p$:

- Se $\alpha \neq 0$, então $\deg(\alpha \cdot p) = \deg(p) \leq d$
- Se $\alpha = 0$, então $\alpha \cdot p$ é o polinômio nulo, cujo grau é $-\infty$ (por convenção, menor que $d$)

Em ambos os casos, $\deg(\alpha \cdot p) \leq d$, logo $\alpha \cdot p \in \mathcal{P}_d$.

**Segunda propriedade verificada.** 


$\mathcal{P}_d$ é um **subespaço linear** de $\mathcal{F}$.

Como $L_X: \mathcal{F} \to \mathbb{R}^m$ é linear (pelo item a) e sua restrição a um subespaço preserva a linearidade, concluímos que **$L_X$ restrita a $\mathcal{P}_d$ também é uma transformação linear**.

## Item c)
**Enunciado:** Para $m = d+1$, considere $\{p_1, \ldots, p_{d+1}\}$ onde $p_i(x_j) = \delta_{ij}$. Mostre que são LI e que formam uma base de $\mathcal{P}_d$.

**Solução:**


Primeiro, vamos conferir se o conjunto $B = \{1, x, \dots, x^d\}$ é realmente uma base de $\mathcal{P}_d$, ou seja, vamos verificar se $B$ é L.I. e se $B$ gera $\mathcal{P}_d$. Como todos os elementos de $B$ estão contidos em $\mathcal{P}_d$, não precisamos nos preocupar se o $\operatorname{span}(B)$ está contido em $\mathcal{P}_d$ (pois $\{1, x, \dots, x^d, x^{d+1}\}$ gera $\mathcal{P}_d$ e é L.I., porém $x^{d+1}$ não está contido em $\mathcal{P}_d$).


### 1. Independência linear de $B$

Dizemos que $B$ é L.I. quando
$$
a_0 \cdot 1 + a_1 x + \dots + a_d x^d = 0 \quad \text{para todo } x \in \mathbb{R} \;\Longrightarrow\; a_0 = a_1 = \dots = a_d = 0.
$$

Para provar isso temos 2 maneiras:

**Maneira 1 (Teorema Fundamental da Álgebra):**  
Seja $p(x) = a_0 + a_1 x + \dots + a_d x^d$. Um polinômio de grau $n$ tem no máximo $n$ raízes distintas (reais ou complexas), a menos que seja o polinômio nulo. Na condição de independência linear, $p(x) = 0$ para **todo** $x$, ou seja, $p$ tem infinitas raízes. Pelo TFA, a única possibilidade é $p$ ser o polinômio nulo, o que só acontece quando $a_0 = a_1 = \dots = a_d = 0$.

**Maneira 2 (derivações sucessivas):**  
Sem perda de generalidade (afinal a condição é para todo $x$), tomamos $x = 0$:
$$
a_0 + a_1 \cdot 0 + \dots + a_d \cdot 0 = 0 \;\Longrightarrow\; a_0 = 0.
$$
Descobrimos que $a_0 = 0$. Voltando a $p(x)$, agora temos $p(x) = a_1 x + a_2 x^2 + \dots + a_d x^d$. Derivamos $p$ e aplicamos $x = 0$:
$$
p'(x) = a_1 + 2a_2 x + \dots + d a_d x^{d-1} \quad\Rightarrow\quad p'(0) = a_1 = 0.
$$
Repetindo o processo, obtemos sucessivamente $a_2 = 0$, $a_3 = 0$, ..., $a_d = 0$. Portanto todos os coeficientes são nulos.


### 2. $B$ gera $\mathcal{P}_d$

Para provar que $\mathcal{P}_d$ é gerado por $B$, vamos definir uma transformação linear
$$
T: \mathbb{R}^{d+1} \to \mathcal{P}_d, \qquad T(a_0, a_1, \dots, a_d) = a_0 + a_1 x + \dots + a_d x^d,
$$
e provar que $T$ é bijetora (injetora e sobrejetora). Isso mostrará que a dimensão de $\mathcal{P}_d$ é $d+1$ e, pelo teorema que diz que todo conjunto de vetores L.I. com o mesmo número de elementos que a dimensão é base, concluiremos que $B$ é base.

- **Injetividade:** Basta mostrar que $\ker(T) = \{(0,\dots,0)\}$. Pelo resultado demonstrado acima (independência linear de $B$), se $T(a_0,\dots,a_d) = 0$ para todo $x$, então $a_0 = \dots = a_d = 0$. Logo $\ker(T) = \{0\}$ e $T$ é injetora.
- **Sobrejetividade:** Todo polinômio $p(x) \in \mathcal{P}_d$ é da forma $p(x) = b_0 + b_1 x + \dots + b_d x^d$, que pode ser obtido como $T(b_0,\dots,b_d)$. Portanto $T$ é sobrejetora.

Como $T$ é bijetora, $\dim(\mathcal{P}_d) = \dim(\mathbb{R}^{d+1}) = d+1$. Assim, $B$ é L.I. e tem cardinalidade igual à dimensão, logo **$B$ é uma base de $\mathcal{P}_d$**.


### O conjunto $\{p_1, p_2, \dots, p_{d+1}\}$ também é base de $\mathcal{P}_d$

Agora vamos provar que $\{p_1, p_2, \dots, p_{d+1}\}$ é base. Defina a matriz de Vandermonde $\mathcal{V} \in \mathbb{R}^{(d+1)\times(d+1)}$ por
$$
\mathcal{V} = 
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1 & x_2 & \cdots & x_{d+1} \\
x_1^2 & x_2^2 & \cdots & x_{d+1}^2 \\
\vdots & \vdots & & \vdots \\
x_1^d & x_2^d & \cdots & x_{d+1}^d
\end{bmatrix}.
$$
Pelo enunciado, cada $p_i$ satisfaz $p_i(x_j) = \delta_{ij}$. Em termos da base canônica de $\mathbb{R}^{d+1}$, isso significa que, se representarmos $p_i$ pelo seu vetor de coeficientes na base canônica de $\mathcal{P}_d$, então $\mathcal{V}$ vezes esse vetor resulta no $i$-ésimo vetor da base canônica $e_i$ (com $1$ na $i$-ésima posição e $0$ nas demais). Ou seja,
$$
\mathcal{V} \cdot \mathbf{p}_i = e_i,
$$
onde $\mathbf{p}_i$ é o vetor de coeficientes de $p_i$ em relação a $\{1, x, \dots, x^d\}$.

### Forma explícita dos polinômios

Sabendo que $p_i(x_j) = \delta_{ij}$ e que as raízes são os pontos $x_j$ com $j \neq i$, podemos escrever:

$$
p_i(x) = \prod_{\substack{j \neq i}} \frac{x - x_j}{x_i - x_j}
$$

Essa expressão garante que:
- $p_i(x_i) = 1$ (todos os fatores do numerador e denominador se cancelam, resultando em $1$);
- $p_i(x_j) = 0$ para $j \neq i$ (pois o fator $(x_j - x_j)$ aparece no numerador, zerando o produto).

Para mostrar que $\{p_1, \dots, p_{d+1}\}$ é L.I., suponha uma combinação linear nula:
$$
\alpha_1 p_1 + \alpha_2 p_2 + \dots + \alpha_{d+1} p_{d+1} = 0 \quad \text{(polinômio nulo)}.
$$
Escrevendo na base canônica, temos $\sum_{i=1}^{d+1} \alpha_i \mathbf{p}_i = \mathbf{0}$. Multiplicando por $\mathcal{V}$ (que é a matriz de Vandermonde).

### Afirmação: Essa matriz de Vandermonde tem posto coluna completo

Pois quando os pontos $x_1, \dots, x_{d+1}$ são distintos, a matriz de Vandermonde $\mathcal{V}$ tem posto completo.

**Prova:** Seja $\alpha = (\alpha_1, \dots, \alpha_{d+1})$ tal que $\mathcal{V}\alpha = 0$. Isso significa que para cada $k = 0, 1, \dots, d$:

$$
\alpha_1 x_1^k + \alpha_2 x_2^k + \cdots + \alpha_{d+1} x_{d+1}^k = 0.
$$

Defina o polinômio $p(x) = \alpha_1 + \alpha_2 x + \cdots + \alpha_{d+1} x^d$. Então $p(x_j) = 0$ para cada $j = 1, \dots, d+1$. Ou seja, $p$ é um polinômio de grau $\leq d$ que possui $d+1$ raízes distintas. A única possibilidade pelo (T.F.A) é $p \equiv 0$, logo $\alpha_1 = \cdots = \alpha_{d+1} = 0$. Portanto as colunas são L.I. e $\operatorname{posto}(\mathcal{V}) = d+1$.
$$
\mathcal{V} \left( \sum_{i=1}^{d+1} \alpha_i \mathbf{p}_i \right) = \sum_{i=1}^{d+1} \alpha_i (\mathcal{V} \mathbf{p}_i) = \sum_{i=1}^{d+1} \alpha_i e_i = (\alpha_1, \dots, \alpha_{d+1})^T = \mathbf{0}.
$$

Portanto $\alpha_1 = \alpha_2 = \dots = \alpha_{d+1} = 0$. Logo os $p_i$ são linearmente independentes.


Como $\{p_1, \dots, p_{d+1}\}$ possui $d+1$ elementos L.I. e $\dim(\mathcal{P}_d) = d+1$, concluímos que **$\{p_1, p_2, \dots, p_{d+1}\}$ também é uma base de $\mathcal{P}_d$**, como queríamos demonstrar.

## Item d)
**Enunciado:** Use os polinômios de Lagrange para mostrar que $L_X$ tem posto completo.

**Solução:**

A matriz que representa $L_X$ na base canônica de $\mathcal{P}_d$ e na base canônica de $\mathbb{R}^m$ é a matriz de Vandermonde $\mathcal{V} \in \mathbb{R}^{m \times (d+1)}$:

$$
\mathcal{V} = 
\begin{bmatrix}
1 & x_1 & x_1^2 & \cdots & x_1^d \\
1 & x_2 & x_2^2 & \cdots & x_2^d \\
\vdots & \vdots & \vdots & & \vdots \\
1 & x_m & x_m^2 & \cdots & x_m^d
\end{bmatrix}.
$$

Sabemos que $\operatorname{rank}(\mathcal{V}) \le \min\{m, d+1\}$. Queremos provar a igualdade.

---

### Caso 1: $m \ge d+1$

Neste caso, $\min\{m, d+1\} = d+1$. Mostramos que as colunas de $\mathcal{V}$ são linearmente independentes exatamente como fizemos nos itens anteriores: supondo uma combinação linear das colunas nula, obtemos um polinômio $p(x)$ de grau $\le d$ que se anula em $m \ge d+1$ pontos distintos. Pelo Teorema Fundamental da Álgebra (ou pelo fato de que um polinômio não nulo de grau $d$ tem no máximo $d$ raízes), $p$ deve ser o polinômio nulo, logo todos os coeficientes são zero. Portanto $\operatorname{rank}(\mathcal{V}) = d+1$.



### Caso 2: $m < d+1$

Agora $\min\{m, d+1\} = m$. Vamos mostrar que as $m$ **linhas** de $\mathcal{V}$ são linearmente independentes (ou equivalentemente, que a matriz tem posto $m$).

Seja uma combinação linear das linhas igual ao vetor nulo em $\mathbb{R}^{d+1}$:

$$
a_1 \cdot (1, x_1, x_1^2, \dots, x_1^d) +
a_2 \cdot (1, x_2, x_2^2, \dots, x_2^d) +
\cdots +
a_m \cdot (1, x_m, x_m^2, \dots, x_m^d) = (0,0,\dots,0).
$$

Igualando coordenada a coordenada, para cada $k = 0, 1, \dots, d$ obtemos:

$$
a_1 x_1^k + a_2 x_2^k + \cdots + a_m x_m^k = 0.
\tag{1}
$$

#### Restringindo às primeiras $m$ equações ($k = 0, 1, \dots, m-1$)

Como $m < d+1$, temos $m$ equações (para $k = 0, \dots, m-1$) que são suficientes para determinar os $a_i$ (pois o sistema é quadrado). Escrevemos esse sistema na forma matricial:

$$
\begin{cases}
a_1 + a_2 + \cdots + a_m = 0 \\
a_1 x_1 + a_2 x_2 + \cdots + a_m x_m = 0 \\
a_1 x_1^2 + a_2 x_2^2 + \cdots + a_m x_m^2 = 0 \\
\quad \vdots \\
a_1 x_1^{m-1} + a_2 x_2^{m-1} + \cdots + a_m x_m^{m-1} = 0
\end{cases}
$$

A matriz desse sistema é $\widetilde{\mathcal{V}} \in \mathbb{R}^{m \times m}$ definida por $\widetilde{\mathcal{V}}_{ij} = x_j^{\,i-1}$ (linha $i$, coluna $j$), ou seja:

$$
\widetilde{\mathcal{V}} = 
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1 & x_2 & \cdots & x_m \\
x_1^2 & x_2^2 & \cdots & x_m^2 \\
\vdots & \vdots & & \vdots \\
x_1^{m-1} & x_2^{m-1} & \cdots & x_m^{m-1}
\end{bmatrix}.
$$

#### OBS: Retornamos ao caso $m \ge d+1$

Esta matriz $\widetilde{\mathcal{V}}$ é uma matriz de Vandermonde **quadrada** de ordem $m$, com pontos $x_1,\dots,x_m$ distintos. Pelo **caso anterior** (com $d' = m-1$ e $m' = m$), sabemos que ela tem posto **linha** completo. Porém, a quantidade de linhas e colunas é a mesma (é inversível). Mais diretamente: as colunas de $\widetilde{\mathcal{V}}$ são linearmente independentes (o mesmo argumento do polinômio de grau $\le m-1$ com $m$ raízes).


Como $\widetilde{\mathcal{V}}$ é inversível, o sistema linear homogêneo $\widetilde{\mathcal{V}} \mathbf{a} = \mathbf{0}$ admite apenas a solução trivial $\mathbf{a} = \mathbf{0}$. Portanto

$$
a_1 = a_2 = \dots = a_m = 0.
$$

Isso mostra que as $m$ linhas de $\mathcal{V}$ são linearmente independentes. Logo $\operatorname{rank}(\mathcal{V}) = m$.


Nos dois casos, $\operatorname{rank}(L_X) = \min\{m, d+1\}$, como queríamos demonstrar.

## Item f)
**Enunciado:** Qual o número de operações para calcular $p_1(x)$? É possível aproveitar para outro $x'$?

**Solução:**

Como vimos no item (c), o polinômio interpolador de Lagrange $p_i(x)$ tem a forma:

$$
p_i(x) = \prod_{\substack{j=1 \\ j \neq i}}^{m} \frac{x - x_j}{x_i - x_j},
$$

onde $m = d+1$ (número de pontos).

Vamos analisar o custo computacional para calcular $p_i(x)$ para um dado $x$.

O denominador é $D_i = \prod_{j \neq i} (x_i - x_j)$. Para calculá-lo:

- São $m-1$ subtrações: $(x_i - x_j)$ para cada $j \neq i$.
- Em seguida, $m-2$ multiplicações para combinar esses $m-1$ fatores.

Total para o denominador: $(m-1) + (m-2) = 2m - 3$ operações.

Este valor $D_i$ pode ser **armazenado** (uma vez calculado) e reaproveitado para calcular $p_i$ em diferentes pontos $x$, pois não depende de $x$.

Já o numerador é $N_i(x) = \prod_{j \neq i} (x - x_j)$. Para um $x$ específico:

- São $m-1$ subtrações: $(x - x_j)$ para cada $j \neq i$.
- Em seguida, $m-2$ multiplicações para combinar esses $m-1$ fatores.

Total para o numerador: $(m-1) + (m-2) = 2m - 3$ operações.


Após calcular numerador e denominador, o valor de $p_i(x)$ é dado por $N_i(x) / D_i$ (ou $N_i(x) \cdot (1/D_i)$). Isso requer **1 multiplicação** (se já tivermos $1/D_i$ armazenado) ou 1 divisão.


### Custo total para o **primeiro** ponto $x$

- Denominador: $2m - 3$ (calculado uma única vez para cada $i$)
- Numerador: $2m - 3$
- Multiplicação final: $1$

Total: $(2m-3) + (2m-3) + 1 = 4m - 5$ operações.


### Custo para um **segundo** ponto $x'$ (aproveitando o denominador armazenado)

Se já calculamos e armazenamos $D_i$ (ou melhor, $1/D_i$), então para o novo ponto $x'$:

- Numerador: $2m - 3$ operações
- Multiplicação por $1/D_i$: $1$ operação

Total: $(2m-3) + 1 = 2m - 2 = 2(m-1)$ operações.

Portanto, não é necessário calcular os coeficientes do polinômio $p_i$ na base canônica para avaliá-lo em pontos. A forma fatorada (produtório) já permite avaliação direta com custo linear em $m$, e o denominador pode ser pré-computado e reaproveitado para múltiplas avaliações.

## Item g)
**Enunciado:** Complexidade da matriz $L_X$ na base canônica vs base de Lagrange.

**Solução:**

Primeiro, lembremos que a matriz de $L_X$ representa a transformação linear que avalia polinômios nos pontos $X = \{x_1, \dots, x_m\}$. Vamos considerar o caso em que $m = d+1$ (número de pontos igual à dimensão de $\mathcal{P}_d$), pois a base de Lagrange foi definida nessa situação.


### 1. Na base canônica $\{1, x, x^2, \dots, x^d\}$

A matriz é a matriz de Vandermonde $\mathcal{V} \in \mathbb{R}^{m \times m}$ com entradas $\mathcal{V}_{ij} = x_i^{\,j-1}$ (linha $i$, coluna $j$).

Para calcular essa matriz, podemos gerar as potências sucessivas para cada ponto $x_i$:

- Começamos com $x_i^0 = 1$ (custo 0).
- Para obter $x_i^1$, fazemos 1 multiplicação: $x_i \cdot 1$.
- Para $x_i^2$, fazemos 1 multiplicação: $x_i \cdot x_i^1$.
- Em geral, cada potência $x_i^k$ (para $k \ge 1$) requer 1 multiplicação a partir da potência anterior.

Assim, para um único ponto $x_i$, precisamos de $d$ multiplicações para gerar $x_i^1, x_i^2, \dots, x_i^d$. Como temos $m = d+1$ pontos, o número total de multiplicações é:

$$
m \cdot d = (d+1) \cdot d = O(d^2).
$$

Não há necessidade de divisões ou outras operações. Portanto, a complexidade para calcular a matriz de $L_X$ na base canônica é **$\mathbf{O(m \cdot d)}$**, que no caso $m = d+1$ é $\mathbf{O(m^2)}$.



### 2. Na base dos polinômios interpoladores de Lagrange $\{p_1, \dots, p_m\}$

Nesta base, cada $p_i$ satisfaz $p_i(x_j) = \delta_{ij}$. Portanto:

$$
L_X(p_i) = (p_i(x_1), \dots, p_i(x_m))^T = e_i,
$$

onde $e_i$ é o $i$-ésimo vetor da base canônica de $\mathbb{R}^m$. Consequentemente, a matriz de $L_X$ na base de Lagrange (domínio) e base canônica (contradomínio) é a **matriz identidade** $I_m$.

Para "calcular" essa matriz, não precisamos realizar nenhuma operação aritmética, pois já conhecemos sua forma. Podemos simplesmente escrever $I_m$ diretamente. Portanto, a complexidade é **$\mathbf{O(1)}$** (ou constante).


### Comparação

| Base | Matriz obtida | Complexidade |
|------|---------------|---------------|
| Canônica | \mathcal{V}andermonde | $O(m^2)$ |
| Lagrange | Identidade | $O(1)$ |

A escolha da base dos polinômios de Lagrange simplifica drasticamente a representação de $L_X$, reduzindo o custo de construção da matriz a zero.
